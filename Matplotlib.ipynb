# DataAnalysis for Telecom Company

```python
#Matplotlib 
import matplotlib.pyplot as mp
import pandas as pd
customer_data=pd.read_csv('customer_churn.csv')
customer_data
```

```python
customer_data.values
```

```python
%matplotlib inline
x=customer_data['Contract'].value_counts().keys().to_list()
y=customer_data['Contract'].value_counts()
mp.bar(x,y,color='orange',edgecolor='black')
mp.title('Distribution of Contract')
mp.xlabel('Contract type of customer')
mp.ylabel('Count')
mp.show()
```

```python
%matplotlib inline
x=customer_data['MonthlyCharges']
mp.hist(x,color='forestgreen',bins=10)
mp.title('Distribution of Monthly charges')
mp.xlabel('Monthly charges incurred')
mp.ylabel('Count')
mp.grid(True, color='black')
mp.show()
```

```python
%matplotlib inline
x=customer_data['tenure'].dropna()
y=customer_data['TotalCharges'].dropna()
mp.scatter(x,y,color='indigo')
mp.title('Total Charges vs Tenure')
mp.xlabel('Tenure of the customer')
mp.ylabel('Total Charges Incurred')
mp.show()
```

```python
%matplotlib inline
customer_data.boxplot(column='MonthlyCharges', by='PaymentMethod', figsize=(10,5))
mp.title('MonthlyCharges vs PaymentMethod')
mp.xlabel('Payment Method of customer')
mp.ylabel('Monthly Charges Incurred')
mp.show()
```
